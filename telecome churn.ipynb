<!-- Using groupby() First, the groupby method divides the grouping_columns by their values. They become a new index in the resulting dataframe. Then, columns of interest are selected (columns_to_show). If columns_to_show is not included, all non groupby clauses will be included. Finally, one or several functions are applied to the obtained groups per selected columns. # Summary tables Suppose we want to see how the observations in our sample are distributed in the context of two variables - Churn and International plan. To do so, we can build a contingency table using the crosstab method:

```python
pd.crosstab(index=df['Churn'],
            columns=df['International plan'],normalize=True)
```

<!-- We can see that most of the users are loyal and do not use additional services (International Plan/Voice mail).

This will resemble pivot tables to those familiar with Excel. And, of course, pivot tables are implemented in Pandas: the pivot_table method takes the following parameters:

values – a list of variables to calculate statistics for, index – a list of variables to group data by, aggfunc – what statistics we need to calculate for groups, ex. sum, mean, maximum, minimum or something else. Let's take a look at the average number of day, evening, and night calls by area code: -->

```python
df.pivot_table(index=['Total day calls','Total eve calls','Total night calls'],
              columns='State',
              aggfunc=np.mean)
```

<!-- DataFrame transformations Like many other things in Pandas, adding columns to a DataFrame is doable in many ways.

<!-- df.insert(...): This function is used to insert a new column into an existing DataFrame. loc=len(df.columns): This argument determines the index (position) where the new column should be inserted. len(df.columns) gives you the total number of columns in your DataFrame. By setting loc to this value, you are inserting the new column at the very end of the DataFrame. column="Total calls": This specifies the name of your new column. value=total_calls: This provides the data that will populate the "Total calls" column. We calculated this data in the previous step. -->

```python
df['Total charges']=df['Total day charge']+df['Total night charge']+df['Total eve charge']+df['Total international charge']
```
